# Aufgabe 2 

a)

$E(mx, my, r) = \sum_{i=1}^{50} \left( \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r \right)^2$

Die Residuen der Kostenfunktion geben die Differenz zwischen dem tatsächlichen Abstand zum Mittelpunkt des Kreises $(mx,my)$ und dem Radius $r$ an. Also: 

$r_i(\theta) = \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r$

Die Kostenfunktion $E(\theta)$ kann also als Summe der quadrierten Residuen interpretiert werden:

$E(\theta) = \sum_{i=1}^{50} (r_i(\theta))^2 = \sum_{i=1}^{50} \left( \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r \right)^2$

Was wiederum die ursprüngliche Kostenfunktion ergibt.

b)

Partielle Ableitung $\frac{\partial r_i}{\partial mx}$ :  

$\frac{\partial r_i}{\partial mx} = \frac{\partial}{\partial mx}\left( \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r \right) \\
= \frac{1}{2\sqrt{(x_i - mx)^2 + (y_i - my)^2}} \cdot 2(x_i - mx) \cdot (-1) \\
= \frac{-(x_i - mx)}{\sqrt{(x_i - mx)^2 + (y_i - my)^2}}$

Partielle Ableitung $\frac{\partial r_i}{\partial my}$:  

$\frac{\partial r_i}{\partial my} = \frac{\partial}{\partial my}\left( \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r \right) \\
= \frac{1}{2\sqrt{(x_i - mx)^2 + (y_i - my)^2}} \cdot 2(y_i - my) \cdot (-1) \\
= \frac{-(y_i - my)}{\sqrt{(x_i - mx)^2 + (y_i - my)^2}}$

Partielle Ableitung $\frac{\partial r_i}{\partial r}$:  

$\frac{\partial r_i}{\partial r} = \frac{\partial}{\partial r}\left( \sqrt{(x_i - mx)^2 + (y_i - my)^2} - r \right) \\
= -1$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle


def compute_jacobian_and_residuals(points, theta):
    mx, my, r = theta
    J = np.zeros((len(points), 3))
    residuals = np.zeros(len(points))

    for i, (xi, yi) in enumerate(points):
        dx = xi - mx
        dy = yi - my
        dist = np.sqrt(dx**2 + dy**2)

        # Residuen
        residuals[i] = dist - r

        # Jakobi-Matrix
        # Einsetzen in die partiellen Ableitungen
        J[i, 0] = -dx / dist  
        J[i, 1] = -dy / dist  
        J[i, 2] = -1  

    return J, residuals
```

```python
def minimize_cost_function(points, initial_theta):
    theta = initial_theta
    num_iterations = 500
    learning_rate = 1e-2

    for _ in range(num_iterations):
        J, residuals = compute_jacobian_and_residuals(points, theta)
        d_theta = np.linalg.lstsq(J, -residuals, rcond=None)[0]
        theta += learning_rate * d_theta
    return theta
```

```python
def plot_results(points, initial_theta, optimized_theta):
    
    fig, ax = plt.subplots()

    ax.scatter(points[:, 0], points[:, 1], label='Datenpunkte')

    initial_circle = Circle((initial_theta[0], initial_theta[1]), initial_theta[2], fill=False, color='red', label='Startkreis')
    ax.add_patch(initial_circle)

    optimized_circle = Circle((optimized_theta[0], optimized_theta[1]), optimized_theta[2], fill=False, color='blue', label='Optimierter Kreis')
    ax.add_patch(optimized_circle)

   
    ax.set_title('Optimierung von Kreisparametern')
    ax.set_aspect('equal', 'box')
    ax.legend()

    plt.show()
    
def generate_points():
    angles = np.random.uniform(0, 2*np.pi, 50)
    noise_x = np.random.normal(0, 0.1, 50)
    noise_y = np.random.normal(0, 0.1, 50)
    
    x = 2 * np.cos(angles) + noise_x
    y = 2 * np.sin(angles) + noise_y
    
    return x, y
```

```python
np.random.seed(0)
points = np.random.rand(50, 2) * 4 - 2
x, y = generate_points()
points = list(zip(x, y))
points = np.array(points)
initial_theta = np.array([0.5, 0.1, 0.9])
optimized_theta = minimize_cost_function(points, [0.5, 0.1, 0.9])
print(initial_theta)


plot_results(points, initial_theta, optimized_theta)
```

# Aufgabe 3

Datenterm:

$\sum_{(i,j) \in \kappa} (r_{ij} - \vec{p}_i^T \vec{q}_j)^2$

```python
def compute_cost(R, P, Q, I, lambda_reg):
    m, n = R.shape 
    reg_term = 0
    
    # Datenterm
    for i in range(m):
        for j in range(n):
            data_term += (R[i, j] - np.dot(P[i, :], Q[j, :]))**2
    
    # Regularisierung
    for i in range(m):
        reg_term += np.linalg.norm(P[i])**2 
        
    for j in range(n):
        reg_term += np.linalg.norm(Q[j])**2

    reg_term *= lambda_reg

    
    total_cost = data_term + reg_term
    
    return total_cost
```

```python
def solveBlockP(Q, R, I, i, λ):
    Ii = np.where(I[i, :] == 1)[0]
    
    # λ * npi * I
    npi = len(Ii)
    reg_term = λ * npi * np.identity(Q.shape[1])
    
    sum_qj_qjT = np.zeros((Q.shape[1], Q.shape[1]))
    sum_qj_rij = np.zeros(Q.shape[1])
    
    for j in Ii:
        qj = Q[j, :]
        sum_qj_qjT += np.outer(qj, qj)
        sum_qj_rij += qj * R[i, j]
    
    pi = np.linalg.inv(sum_qj_qjT + reg_term) @ sum_qj_rij
    return pi

def solveBlockQ(P, R, I, j, λ):
    Ij = np.where(I[:, j] == 1)[0]
    
    # λ * nqj * I
    nqj = len(Ij)
    reg_term = λ * nqj * np.identity(P.shape[1])
    
    sum_pi_piT = np.zeros((P.shape[1], P.shape[1]))
    sum_pi_rij = np.zeros(P.shape[1])
    
    for i in Ij:
        pi = P[i, :]
        sum_pi_piT += np.outer(pi, pi)
        sum_pi_rij += pi * R[i, j]
    
    
    qj = np.linalg.inv(sum_pi_piT + reg_term) @ sum_pi_rij
    return qj
```

```python
from utility import *

def compute_data_term(R, P, Q, I):
    data_term = 0
    num_observations = 0
    m, n = R.shape
    for i in range(m):
        for j in range(n):
            if I[i, j] == 1:
                error = R[i, j] - np.dot(P[i, :], Q[j, :])
                data_term += error ** 2
                num_observations += 1
    return data_term, num_observations

def compute_rmse(data_term, num_observations):
    rmse = np.sqrt(data_term / num_observations)
    return rmse

def als_algorithm(R, I, num_factors, lambda_reg, num_iterations):
    m, n = R.shape
    P = np.random.rand(m, num_factors)
    Q = np.random.rand(n, num_factors)  
    
    for iteration in range(num_iterations):
        for i in range(m):
            P[i, :] = solveBlockP(Q, R, I, i, lambda_reg)
        

        for j in range(n):
            Q[j, :] = solveBlockQ(P, R, I, j, lambda_reg)
        
        data_term, num_observations = compute_data_term(R, P, Q, I)
        rmse = compute_rmse(data_term, num_observations)
        print(f'Iteration: {iteration}, RMSE: {rmse}')
    
    return P, Q, rmse

R, I = setup_dataset(toy_ds=False)  
num_factors = 10  
lambda_reg = 0.1  
num_iterations = 20 

P, Q, rmse = als_algorithm(R, I, num_factors, lambda_reg, num_iterations)
```
